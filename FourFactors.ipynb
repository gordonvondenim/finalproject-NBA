```python
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score
scores = "team_boxscore_stats_1997-19.csv"
```

```python
scores_df = pd.read_csv(scores)
scores_df['WL'] = scores_df['WL'].map({'W': 1, 'L': 0})
scores_df.head()
```

```python
df_2018 =scores_df[49678:]
df_2018.head()
```

```python
ratings_2018 = '2019_team_stats.csv'
```

```python
ratings_2018 = pd.read_csv(ratings_2018)
ratings_2018.head()
```

```python
ratings_2018 = ratings_2018.rename(columns=ratings_2018.iloc[0]).drop(ratings_2018.index[0])
```

```python
merged_df = pd.merge(df_2018, ratings_2018, left_on='TEAM_NAME', right_on='Team')
merged_df.head()
```

```python
y = merged_df['WL'].dropna().values.reshape(-1,1)
x = merged_df[['OeFG%','OTOV%','ORB%','OFT/FGA','DeFG%','DTOV%','DRB%','DFT/FGA']].astype(float)
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
model = LogisticRegression()
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y)

model.fit(X_train,y_train)
```

```python
trainz = model.score(X_train, y_train)
testz = model.score(X_test, y_test)
print(trainz,testz,trainz-testz)
```
